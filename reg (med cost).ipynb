```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

plt.figure(figsize = (30,50))
plotnumber = 1

for column in data:
    if plotnumber <= 30:
        ax = plt.subplot(6, 2, plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column)
    plotnumber += 1

plt.tight_layout()
plt.show
```

```python
def plot_correlation_map(data):
    corr = data.corr()
    s, ax = plt.subplots(figsize = (20,12))
    cmap = sns.diverging_palette(220,120,as_cmap = True)
    s = sns.heatmap(
        corr,
        cmap = cmap,
        square = True,
        cbar_kws = {'shrink' : 1.2},
        ax = ax,
        annot = True,
        annot_kws = {'fontsize' : 12},    
    )
plot_correlation_map(data)
```

```python
df_sum = data.describe()
```

```python
def draw_axvlines(plt, col):
    mean = df_sum.loc["mean", col]
    q1 = df_sum.loc["25%", col]
    q2 = df_sum.loc["50%", col]
    q3 = df_sum.loc["75%", col]
    plt.axvline(mean, color = 'g')
    plt.axvline(q1, color = "b");               
    plt.axvline(q2, color = "navy");            
    plt.axvline(q3, color = "purple");           
    plt.legend({"Mean": mean, "25%" : q1, "50%" : q2, "75%" : q3});

fig, axes = plt.subplots(3, 2, figsize = (20, 12));
fig.suptitle('Distribution charts for Age, bmi and smokers.'); 

sns.boxplot(data["age"], ax = axes[0][0]);
axes[0][0].set(xlabel = 'age');

pp = sns.distplot(data["age"], ax = axes[0][1], bins = 10);
axes[0][1].set(xlabel = 'Distribution of Age');
draw_axvlines(pp, "age");

sns.boxplot(data["bmi"], ax = axes[1][0], 
            color = "mediumslateblue");
axes[1][0].set(xlabel = 'Distribution of bmi');

pp = sns.distplot(data["bmi"], ax = axes[1][1], bins = 10, 
                  color = "mediumslateblue");
axes[1][1].set(xlabel = 'Distribution of bmi');
draw_axvlines(pp, "bmi")

sns.boxplot(data["smoker"], ax = axes[2][0], color = "mediumslateblue");
axes[2][0].set(xlabel = 'Distribution of smoker');

pp = sns.distplot(data["smoker"], ax = axes[2][1], 
                  color = "mediumslateblue");
axes[2][1].set(xlabel = 'Distribution of smoker');
draw_axvlines(pp, "smoker")
```

# Linear reg

```python
# defining independent and dependent variables
X=data.iloc[:,0:-1] # independent variables
y=data.charges  # dependent variable
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
l_reg = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, random_state = 42)
```

```python
l_reg.fit(X_train, y_train)
l_reg.intercept_, l_reg.coef_
```

```python
y_pred = l_reg.predict(X_test)
```

```python
y_pred
```

## Eval: linear reg

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

test_score=r2_score(y_test, y_pred)
print(test_score)
```

```python
## mean squared error and mean absolute error of testing data

print('MSE',mean_squared_error(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
## Root mean squared error

import math
print(math.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
## adjusted r2 score

train_adjusted_r2score=1-(1-0.71)*(1003-1)/(1003-6-1)
train_adjusted_r2score
```

```python
test_adjusted_r2score=1-(1-0.74)*(335-1)/(335-6-1)
test_adjusted_r2score
```

# Eval: Decision tree

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(criterion = 'mse', splitter = 'best', random_state = 42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
y_pred_dt
```

```python
print('mse ', mean_squared_error((y_test), (y_pred_dt)))
print('mae ', mean_absolute_error((y_test), (y_pred_dt)))
```

# Hyperperameter for decision tree

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters={"splitter":["best","random"],
            "max_depth" : [1,3,5,7,9],
           "min_samples_leaf":[1,2,3,4,5,],
           "max_features":["auto","sqrt"],
           "max_leaf_nodes":[5,10,15] }
```

```python
hyper_tuning = RandomizedSearchCV(estimator = dt, param_distributions = parameters, cv = 2, n_iter = 10, n_jobs = -1)
```

```python
hyper_tuning.fit(X_train, y_train)
```

```python
hyper_tuning.best_params_
```

```python
dt_hyper = DecisionTreeRegressor(splitter = 'best', min_samples_leaf = 4, max_leaf_nodes = 15, max_features = 'auto', max_depth = 7)
```

```python
dt_hyper.fit(X_train, y_train)
```

```python
tuning_pred = dt_hyper.predict(X_test)
```

```python
tuning_pred
```

```python
print('Decision Tree Regressor hyperparameter score testing :', hyper_tuning.score(X_test,y_test))
```
